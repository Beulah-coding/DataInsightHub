# Basic Statistics

#### Statistics provides tools and methods to find structure and to give deeper data insights.

**Mean, Mode and Median**

*Mean*: Also known as average or Expectation provides a central value of a population (a number of observations). 
*Mode*: The most occurence in an observation.
*Median*: The mid observation of a sorted population

```python
import numpy as np
import statistics as stats

pop = np.array([1, 4, 5, 6, 8, 9, 9, 4, 10])

mean_pop = np.mean(pop)
mode_pop = stats.mode(pop)
median_pop = stats.median(pop)

print("The mean of the observation {} is {}".format(pop, round(mean_pop, 2)))
print("The mode of the observation {} is {}".format(pop, mode_pop))
print("The median of the observation {} is {}".format(pop, median_pop))
```

**Variance and Standard Deviation**

*Variance*: Measures how much the data points are spread out from the mean.
*Standarad Deviation*: It is often preferred over the variance because it has the same unit as the data points, which means you can interpret it more easily

```python
import numpy as np

pop = np.array([1, 4, 5, 6, 9, 9])

var_pop = np.var(pop)
std_pop = np.std(pop)

print("The variance of the observation {} is {}".format(pop, round(var_pop, 2)))
print("The standard deviation of the observation {} is {}".format(pop, round(std_pop, 2)))
```

**Covariance and Corelation**

*Covariance*: measures the joint variability of two random variables and describes the relationship between these two variables.
*Correlation*: in addition, it also measures both the strength and the direction of the linear relationship between two variables.


```python
import numpy as np

pop1= np.array([3, 6, 8, 9])
pop2 = np.array([-1,-4,-7,-8])

cov_pop1_2 = np.cov(pop1, pop2)
cor_pop1_2 = np.corrcoef(pop1, pop2)

print("Covariances:", cov_pop1_2, "\n")
print("Correlation:", cor_pop1_2)
```

####  Probability Distribution Function (PDF) or Probability Density

**Bionomials Distribution**: A discrete propability distribution.

```python
# Histogram of binomials using Random Generation of 1000 independent samples
import numpy as np
import matplotlib.pyplot as plt
n = 7
p = 0.13
N = 1000

rand_pop = np.random.binomial(n,p,N)
counts, bins, ignored = plt.hist(rand_pop, 15, density = True, rwidth = 0.8, color = 'orange')

plt.title("Binomial distribution with p = 0.13 n = 7")
plt.xlabel("Number of successes")
plt.ylabel("Probability")
plt.show()
```

**The Poisson distribution**: A discrete propability distribution

```python
# Histogram of Poisson distributionusing Random Generation of 1000 independent Binomial samples
import numpy as np

lambda_ = 6
N = 1000
X = np.random.poisson(lambda_,N)

counts, bins, ignored = plt.hist(X, 50, density = True, color = 'Blue')
plt.title("Randomly generating from Poisson Distribution with lambda = 6")
plt.xlabel("Number of visitors")
plt.ylabel("Probability")
plt.show()
```

**Normal probability distribution**: is the continuous probability distribution for a real-valued random variable

```python
# Histogram of Normal distribution
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt


mu = 0
sigma = 1
N = 1000
norm_pop = np.random.normal(mu,sigma,N)
x_values = np.arange(-5,5,0.01)
y_values = norm.pdf(x_values)

counts, bins, ignored = plt.hist(norm_pop, 30, density = True,color = 'Green',label = 'Sampling Distribution')
plt.plot(x_values,y_values, color = 'r',linewidth = 2.5,label = 'Population Distribution')
plt.title("Randomly generating 1000 obs from Normal distribution mu = 0 sigma = 1")
plt.ylabel("Probability")
plt.legend()
plt.show()
```
